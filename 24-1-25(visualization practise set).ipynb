```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,color='yellow',linestyle='dashed',linewidth=2,marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['jan','feb','march','april','may','june']
y=[3000,4000,5000,3000,6000,8000]
plt.plot(x,y,color='pink',linestyle='dashed',marker='*')
plt.xlabel('months')
plt.ylabel('savings amount')
plt.title('line plot 2')
```

```python
import numpy as np

categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['yellow','lavender','pink','skyblue','orange'])
plt.xlabel('categories')
plt.ylabel('values')
plt.title('bar chart')

plt.show()
```

```python
# bins : defines the intervals for grouping of data
ages=[20,22,23,25,27,27,28,30,31,32,34,35,35,40,58,45,67]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='peachpuff',edgecolor='black')
plt.title("Age distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ['Python', 'C++', 'Java', 'Ruby']
percentages = [40.0, 20.0, 30.0, 10.0]



plt.pie(percentages, labels=subjects, autopct='%1.1f%%',startangle=90,colors=['lightgreen', 'lavender', 'peachpuff', 'lightcoral'])


plt.title('Subject Percentages')

plt.legend()


plt.show()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='lightgreen',s=500,edgecolor='black',marker='*')
plt.title("scatter plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

```
